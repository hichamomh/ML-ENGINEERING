# Machine Learning vs Rule-Based Systems 1.2

thumbnail-1-02.jpg

Imagine you want to detect if a message is spam or not. How would you do it?

**Rule-based systems**

* For rule-based systems you need write a program or bunch of codes to catch if the eamil is spam or not.

* You have to write code such as if the mail is from this specific sender, if it contains a certain word or not.

* This is so inefficient, because the scammers will keep changing the text in the images.

* More headache to the developer. 🤷‍♀️🤷‍♂️


**Machine Learning**
--------------
So, how do you avoid all these stress with machine learning.

* Basically, you just need to create simple rules to test if a mail is spam or not.

* So, you could create rules of is the email coming from the kind of mail, represent yes as 1 and no as 0. These different "tests" can be called features.

* So, we could have a table of features from different features you feel should make a mail spam.

* Then, give the users access to label as spam or not. This is really common for email providers

* Put the data into a machine learning (statistical) model to predict if it is spam

- What do you think will be result of the model?

* Definitely the model will be a number, but this time around a probability between 0 and 1. It could be 0.45, 0.94, 0.5, 0.8

* The next thing is to create a threshold, this is a function that rounds result.

* You could decide that if the threshold is 0.5,every probability equal to or above 0.5 is seen as 1

* This is a classification example. 1 could be denoted as Spam, while 0 is denoted as "Not Spam"
